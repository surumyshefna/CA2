1.
Determine the distribution of employees across each team and calculate the percentage split relative to the total number of employees.

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
team=com["Team"].value_counts()
total_employees = len(com)
percentage_split = (team / total_employees) * 100
percentage_split
```

```python
team
```

```python
team.describe()
```

```python
plt.pie(team,labels=team.index,autopct="%1.1f%%",)
plt.title("Employee Distribution by team")
plt.show()
```

```python
sns.histplot(data=team,kde=True)
plt.fonts={"family":"monospace",'color':"g","size":25}
plt.title("TEAM-MEMBER GRAPH")
plt.xlabel("No. OF MEMBERS",loc="center")
plt.ylabel("No. OF TEAMS",loc="center")
plt.show()
```

Minnesota Timberwolves ' has lowest no of employees that is 14.

Generally we can say that almost team got around 15 employees in their team.

 ? Discover which team and position have the highest salary expenditure.

? Investigate if there's any correlation between age and salary, and represent it visually.

```python
sns.scatterplot(x='Age',y='Salary',data=com)
plt.title('Correlation between Age - Salary')
plt.show()
```

Most of the employees aged between 24 and 32 years got 1M+ salary than that of others. Age below 24 and above 32 were getting around 1M salary, which indicates most of the freshers and the mid-aged may be less efficient than that of others.
